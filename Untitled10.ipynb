```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset

df = pd.read_csv("Titanic-Dataset.csv")

# Display first 5 rows
df.head(7)
```

```python
df.dtypes
```

```python
print(df.isna().sum())
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna("Unknown", inplace=True)
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

```python
df.describe(include='all')
```

```python
# Plot histogram for Age distribution
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Plot histogram for Fare distribution
plt.hist(df['Fare'].dropna(), bins=30, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Scatter plot: Age vs Fare
plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

# Scatter plot: Age vs Survival (to check correlation with survival)
plt.scatter(df['Age'], df['Survived'])
plt.title("Age vs Survival")
plt.xlabel("Age")
plt.ylabel("Survived")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[['Age', 'Fare', 'Survived']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Survival Rate by Passenger Class
pclass_survival = df.groupby('Pclass')['Survived'].mean()

# Plot Survival Rate by Class
pclass_survival.plot(kind='bar', figsize=(6,4))
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival Rate by Gender
sex_survival = df.groupby('Sex')['Survived'].mean()

# Plot Survival Rate by Gender
sex_survival.plot(kind='bar', figsize=(6,4))
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```
